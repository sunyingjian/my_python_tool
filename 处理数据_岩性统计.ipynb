```python
import numpy as np
import pandas as pd
```

```python
lithology_numbers = {30000: 0,
                 65030: 1,
                 65000: 2,
                 80000: 3,
                 74000: 4,
                 70000: 5,
                 70032: 6,
                 88000: 7,
                 86000: 8,
                 99000: 9,
                 90000: 10,
                 93000: 11}
```

```python
result = pd.DataFrame()
```

```python
result
```

```python
import glob
path = r'fenjing'
allFiles = glob.glob(path + "/*.csv")
```

```python
for file_ in allFiles:
    df = pd.read_csv(file_)
    y = df['FORCE_2020_LITHOFACIES_LITHOLOGY']
    y = y.map(lithology_numbers)
    df['FORCE_2020_LITHOFACIES_LITHOLOGY'] = y
    a = df['FORCE_2020_LITHOFACIES_LITHOLOGY'].value_counts()
    temp = []
    for i  in range(12):
        try:
            temp.append(a[i])
        except:
            temp.append(0)
    result[file_[8:-4]] = temp
```

```python
result = result.T
```
